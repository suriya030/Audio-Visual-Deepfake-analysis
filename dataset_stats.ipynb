```python
import cv2
import os
from collections import defaultdict
from tqdm import tqdm
```

```python

# Path to the folder containing videos
# video_folder = r"C:\Users\Admin\Desktop\cs229_project\train\video\0_real"
video_folder = r"C:\Users\Admin\Desktop\cs229_project\train\video\1_fake"
```

```python
# Dictionaries to store counts
frame_count_dict = defaultdict(int)
fps_dict = defaultdict(int)

# Iterate over all video files in the folder
for filename in tqdm(os.listdir(video_folder)):
    if filename.endswith((".mp4", ".avi", ".mov", ".mkv", ".flv", ".wmv")):
        video_path = os.path.join(video_folder, filename)
        cap = cv2.VideoCapture(video_path)

        if cap.isOpened():
            frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
            fps = int(cap.get(cv2.CAP_PROP_FPS))

            frame_count_dict[frame_count] += 1
            fps_dict[fps] += 1

        cap.release()


```

```python
# Print results
print("Number of videos with particular frame counts:")
for frame_count, count in sorted(frame_count_dict.items()):
    print(f"Frames: {frame_count} -> Videos: {count}")

print("\nNumber of videos with particular FPS:")
for fps, count in sorted(fps_dict.items()):
    print(f"FPS: {fps} -> Videos: {count}")
```
